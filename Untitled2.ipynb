```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bBooks = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
```

```python
soupOfABook = BeautifulSoup(bBooks, 'html.parser')
```

```python
author = []
topic = []
price = []
pages = []

for book in soupOfABook.find('table').find_all('tr',class_='book'):
  author.append(book.find_all('td')[2].text)
  topic.append(book.find_all('td')[5].text)
  price.append(int(book.find_all('td')[6].text.replace('$','').replace('.','')))
  pages.append(int(book.find_all('td')[4].text))
print(author)
```

```python
table = pd.DataFrame({
    'author':author,
    'topic':topic,
    'prices':price,
    'pages':pages
})
table.info()
print(table)
```

```python
plt.figure(figsize=(16,8))
sns.countplot(x='author',data=table)
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.countplot(x='topic',data=table)
plt.show()
```

```python
pageAvg = table.groupby('topic').agg({'pages':'mean'})
pageAvg
```

```python
priceAvg = table.groupby('topic').agg({'prices':'mean'})
priceAvg
```
